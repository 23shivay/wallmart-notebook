## **Model Building part**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
```

```python
df['Store'].unique()
```

```python
# selecting the 7 stores
stores = [1, 2, 3, 4, 5, 6, 7]
```

```python
# function to forecast using SARIMA
def forcasting_method(store, df, seasonal_period):
    print("\n\n\nForecasting for Store: ",store)

    store_df = df[df['Store'] == store][['Date', 'Weekly_Sales']]
    store_df = store_df.set_index('Date')
    store_df = store_df.resample('W').sum()  # Resample to weekly data

    # train test split
    train_size = int(len(store_df) * 0.8)
    train, test = store_df[:train_size], store_df[train_size:]

    # fitting the SARIMA model
    model = SARIMAX(
        train['Weekly_Sales'],
        order=(1, 1, 1),  # Non-seasonal ARIMA parameters (p, d, q)
        seasonal_order=(1, 1, 1, seasonal_period),  # Seasonal parameters (P, D, Q, m)
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    model_fit = model.fit(disp=False)

    # forecasting process
    steps = len(test)
    forecast = model_fit.forecast(steps=steps)

    # forecasted values with corresponding dates
    forecast_df = pd.DataFrame({
        'Date': test.index,
        'Weekly_Sales_Forecasted': forecast})
    print("\n\nForecasted Values for Store", store," for 12 weeks are as follows:")
    print(forecast_df,'\n\n')

    # comparing forecasted and test data
    plt.figure(figsize=(10, 6))
    plt.plot(train['Weekly_Sales'], label='Train', color='blue')
    plt.plot(test['Weekly_Sales'], label='Test', color='green')
    plt.plot(test.index, forecast, label='Forecast', color='red')
    plt.title(f'Store {store} - Forecast vs Actual')
    plt.xlabel('Date')
    plt.ylabel('Weekly Sales')
    plt.legend()
    plt.show()

    # evaluation of model
    mse = mean_squared_error(test['Weekly_Sales'], forecast)
    print("\n\nMSE for Store ",store,": ",mse,'\n\n')
    return

seasonal_period = 52   # as 52 weeks in an year are there, yearly seasonaility it is refering to

# forecasting for each of the 7 stores selected
for store in stores:
    forcasting_method(store, df, seasonal_period)
```
